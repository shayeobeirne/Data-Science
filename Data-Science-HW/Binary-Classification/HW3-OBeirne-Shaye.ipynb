tripadvisor_review.csv contains some data from a travel reviews dataset, which was created by crawling the traveler’s review from TripAdvisor.com. Reviews on destinations across East Asia are considered in 10 different categories. Each traveler’s rating is categorized from 0(terrible) to 4(excellent), and we have access to average rating. Use the clustering method learned in class to find those with the most similar ratings.

Below, provide your code used to cluster the tripadvisor_review.csv data.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

ratings_ = np.genfromtxt('tripadvisor_review.csv', skip_header = 1, delimiter = ',')
ratings = ratings_[1:,1:]

kmeans = KMeans(n_clusters = 5)
kmeans.fit(ratings)

centroids = kmeans.cluster_centers_
labels = kmeans.labels_

group_zero = []
group_one = []
group_two = []
group_three = []
group_four = []

for i in range(len(labels)):
    name = "User " + str(i+1)
    if labels[i] == 0:
        group_zero.append(name)
    elif labels[i] == 1:
        group_one.append(name)
    elif labels[i] == 2:
        group_two.append(name)
    elif labels[i] == 3:
        group_three.append(name)
    elif labels[i] == 4:
        group_four.append(name)

print("Group A:", group_zero)
print("")
print("Group B:", group_one)
print("")
print("Group C:", group_two)
print("")
print("Group D:", group_three)
print("")
print("Group E:", group_four)
```
